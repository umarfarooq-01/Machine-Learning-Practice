```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv(r"C:\Users\farooq\Downloads\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Capita income(in USD$)')
plt.scatter(df.year,df[['per capita income (US$)']],color='Blue',marker='.')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df[['per capita income (US$)']])
```

```python
reg.predict([[2020]])
```

```python
d= pd.DataFrame({'year':[2017,2018,2019,2020,2021,2022,2023,2024,2025]})
```

```python
p=reg.predict(d)
p
```

```python
d['per capita income (US$)'] = p
```

```python
d
```

```python
final_data=df.merge(d,how='outer',on=['year','per capita income (US$)'])
```
