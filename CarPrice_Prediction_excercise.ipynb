```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv(r"C:\Users\farooq\Downloads\carprices.csv")
df
```

```python
d=pd.get_dummies(df)
d
```

```python
d=d.drop('Car Model_Mercedez Benz C class',axis='columns')
```

```python
d
```

```python
x=d[['Mileage','Age(yrs)','Car Model_Audi A5','Car Model_BMW X5']]
x
```

```python
y=d['Sell Price($)']
y
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
